```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 讀取圖像並轉換為 float64
img = cv2.imread('image_1.png').astype(np.float64)
bgimg = cv2.imread('image_2.png').astype(np.float64)

# 將 BGR 轉為 RGB（因為 matplotlib 顯示的是 RGB）
img_rgb = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_BGR2RGB)
bgimg_rgb = cv2.cvtColor(bgimg.astype(np.uint8), cv2.COLOR_BGR2RGB)

# 取得圖像尺寸
row, col, _ = img.shape

# 建立輸出影像
outputimg = np.zeros((row, col), dtype=np.float64)
binaryimg = np.zeros((row, col), dtype=np.uint8)

# 顯示影像
plt.subplot(1, 2, 1)
plt.imshow(img_rgb.astype(np.uint8))
plt.title('img_1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(bgimg_rgb.astype(np.uint8))
plt.title('img_2')
plt.axis('off')
```

```python
T = 10

# 計算每個像素的距離（用歐氏距離公式）
for i in range(row):
    for j in range(col):
        diff = np.sqrt(
            (bgimg[i, j, 0] - img[i, j, 0]) ** 2 +
            (bgimg[i, j, 1] - img[i, j, 1]) ** 2 +
            (bgimg[i, j, 2] - img[i, j, 2]) ** 2
        )
        outputimg[i, j] = diff
        if diff > T:
            binaryimg[i, j] = 255
```

```python
plt.subplot(1, 2, 1)
plt.imshow(outputimg, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(binaryimg, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(2, 2, 1)
plt.imshow(img_rgb.astype(np.uint8))
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(bgimg_rgb.astype(np.uint8))
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(outputimg, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(binaryimg, cmap='gray')
plt.axis('off')
```
